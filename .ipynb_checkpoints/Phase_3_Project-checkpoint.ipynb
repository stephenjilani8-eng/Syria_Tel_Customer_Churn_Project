**1. BUSINESS UNDERSTANDING**

The telecommunications industry in Syria is highly dynamic, with an oligopolistic market dominated by major competitors MTN Syria and NewComer Wafa Telecom. SyriaTel has been facing a challenge: customers don't stay long and end up switching to competitors.

**BUSINESS OBJECTIVES**
The primary objective is to ascertain if there are any predictable trends in the data that can help the company make informed decisions. The specific objectives are as follows
1. To identify any patterns indicated by the data as far as customers leaving the company is concerned.
2. To ascertain if the patterns indicate particular reasons why customers don't stay for long with SyriaTel.
3. To provide recommendations on what can be done differently to ensure that customers stay for longer.

**BUSINESS QUESTIONS**
1. Are there any trends in the data presented?
2. Are the trends predictive of any possible scenarios?

The project is expected to be a success if there's certainty that there are patterns in the data presented or not; and if patterns exist, that recommendations are provided to management on a way forward to ensure that decisions are made so that less money is lost through customer churn.

```python
#Adding a column on average charge per minute to see if there is consistency on how customers are charged
df['average charge'] = df['total charge']/df['total minutes']
df['average charge'].head()
```

```python
df.head()
```

```python
#Dropping unnecesarry columms

dropped_df = df.drop(columns =['state','area code','number vmail messages'])
dropped_df.head()
```

```python
#Converting international plan columnn and voice mail plan columns to 0 ad 1 in stead of True and False for easier modelling.
dropped_df['international plan'] =dropped_df['international plan'].map({'yes':1,'no':0})
dropped_df['voice mail plan'] =dropped_df['voice mail plan'].map({'yes':1,'no':0})
dropped_df.head()
```

DATA MODELLING

```python
#Selecting columns that are relevant for our modelling

cols = ['account length', 'international plan', 'voice mail plan', 'customer service calls', 'churn','total minutes', 'total charge','average charge']
final_dataset = dropped_df[cols]
final_dataset.head()

#Colmuns like phone number have not been put as part of the variables due to being for descriptive purposes
#Columns about minutes abd charge have been replaced bt the totals and average columns
#The number of phone calls made by a customer has been considered to not have much impact on customer churn
```

```python
#Feature selection; This is where we select what is our X and what is our y
#This is to separate our dependent variable 'churn' from the rest of the columns 
X = final_dataset.drop(columns=['churn'])
y = final_dataset['churn']
```

```python

#First split:Train and validation vs Test
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X, y, test_size=0.2, random_state=7, stratify=y
)

#Second split: Train vs validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.2, random_state=7, stratify=y_train_val
)
```

```python
#To prevent class imbalance due to the churn clients being fewer than the no churn clients
y.value_counts(normalize=True)
```

```python
#Scaling features. This is to ensure that all the numerical values are comparable

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python
#Building a dummy model to see if our model can give usa baseline for comparison

from sklearn.dummy import DummyClassifier
dummy_model = DummyClassifier(strategy='most_frequent')
dummy_model.fit(X_train,y_train)
dummy_model.score(X_val, y_val)
```

```python
from sklearn.metrics import classification_report

y_dummy = dummy_model.predict(X_val)
print(classification_report(y_val, y_dummy))

#The dummy model predicts 0 churns and hence our real model has to outperform this. We will now build our first real model and see the results.
```

```python
#Building our first model which is a Logistic Regression model
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_scaled, y_train)
```

```python
#Evaluation our model to see if it outperforms our dummy model.
y_pred = log_model.predict(X_val_scaled)
print(classification_report(y_val, y_pred))
```

The model above outperforms the dummy model as the recall for Churns is 0.23 compared to 0.00 for the dummy model. This is to say that our model predicts 23% of actual churn clients.

The model also has a precision of 0.64 for churn clients compared to a precision of0.00 in the dummy model. This is to say out of the clients that the model has flagged, 64% actually are churn clients.

```python
#Improvement of the baseline model which is the Logistic Regression
#We will do a Decision Tree as our improved model
from sklearn.tree import DecisionTreeClassifier

d_tree = DecisionTreeClassifier(max_depth=5, random_state=7)
d_tree.fit(X_train, y_train)
```

```python
#Making predictions using the decision tree model above
y_pred_tree = d_tree.predict(X_val)


#Get classification report for the Decision Tree model

print(classification_report(y_val, y_pred_tree))
```

The model comapred to the basic decision tree model has a recall rate of 69% on the churn customers compared to 0.23; it also has a precision of 95% compared to 64%. The f1 score has also improved from 34% to 80%. This is an indication that it surpasses the basic logistic regression model built above.

```python
#Improving on the decision tree model through a random forest model

from sklearn.ensemble import RandomForestClassifier

rand_forest = RandomForestClassifier(n_estimators=100, random_state=7)
rand_forest.fit(X_train, y_train)
```

```python
#Making predictions using the Random Forest Model Above

y_pred_rand_forest = rand_forest.predict(X_val)

#Running the classification report for the model
print(classification_report(y_val, y_pred_rand_forest))
```

The Random Forest and Decision Tree models have the same precision, f1 score, and recall rates. This is to say that these metrics only are not enough to decide which model is more accurate than the other. However, Random forest is known to be the model that is more suitable for real world situations due to its stability and ability to interpret unseen situations better. For that reason, Random Forest was chosen to be the best model.
